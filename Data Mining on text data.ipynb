```python
import nltk
import pandas as pd
from lxml import etree
```

# Data Collection

```python
# Collect the training data file
train = pd.read_csv("train.csv")

# Ids of the XML files
ids = []
# Texts contained in each XML file
texts = []
# Labels associated to the XML file
labels = []

for sample in train.values:
    ids.append(sample[0])
    # For each XML file referenced in train.csv, collect the text contained in it
    xml_file = etree.parse("data/"+sample[1])
    xml_text = xml_file.xpath('/BODY')[0].text or ''
    texts.append(xml_text)
    # The label
    labels.append(sample[2])
    
news_data = pd.DataFrame()
news_data['id'] = ids
news_data['text'] = texts
news_data['label'] = labels
news_data
```

# Frequently mentioned words in business opportunities

```python
# Collecting the business opportunities
indicesToKeep = news_data['label'] == 1
business_opportunities = news_data[indicesToKeep]
business_opportunities = business_opportunities.drop(columns='label')
business_opportunities
```
